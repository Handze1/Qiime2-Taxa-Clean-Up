```python
import pandas as pd
import numpy as np
import regex
```

```python
data=pd.read_csv('taxa.tsv',sep='\t')
feature_table = pd.read_csv("biomtxt/table.from_biom.tsv", sep='\t')
data
```

```python
data = data.drop(["Confidence"], axis=1)
data = data.rename(columns={'Feature ID': 'OTUID'})
data = data.set_index('OTUID')
diff_appr = data
diff_appr
```

```python
diff_appr
diff_appr['Taxon'].replace({r";": ',' }, inplace= True, regex = True)
lst = []
for i in diff_appr.iloc[0:,0]:
    new1 = i.split(",")
    if len(new1) == 7:
        del new1[6]
        if new1[5] == ' g__':
            del new1[5]
            if new1[4] == ' f__':
                del new1[4]
                if new1[3] == ' o__':
                    del new1[3]
                    if new1[2] == ' c__':
                        del new1[2]
                        if new1[1] == ' p__':
                            del new1[1]
                            if new1[0] == ' k__':
                                del new1[0]
    lst.append(new1[-1])
diff_appr["result"] = lst
#diff_appr
```

```python
new = pd.merge(diff_appr, feature_table, on="OTUID")
new
```

```python
new = new.groupby('result').sum()
new = new.reset_index()
new = new.rename(columns={'result': 'Taxa'})
new.to_csv("out.csv", index=True)
```
